Simple script that joins the four csv files downloaded from the FRED into a single dataframe and exports that dataframe to the project folder as a csv

```python
import pandas as pd
from pathlib import Path
```

```python
#importing PCE csv files from project folder
folderpath = 'C:\\Users\\micha\\Desktop\\File References\\CIS480\\Project\\Data'

PCE = pd.read_csv(folderpath + '\\Personal_Consumption_Expenditures_2022-2025.csv')
PCEDG = pd.read_csv(folderpath + '\\Personal_Consumption_Expenditures_Durable_Goods_2022-2025.csv')
PCES = pd.read_csv(folderpath + '\\Personal_Consumption_Expenditures_Services_2022-2025.csv')
PCEF = pd.read_csv(folderpath + '\\Personal_Consumption_Expenditures_Food_2022-2025.csv')

PCEF.rename(columns={'DFXARC1M027SBEA': 'PCEF'}, inplace=True)
```

```python
print(PCE.head())
print(PCEDG.head())
print(PCES.head())
print(PCEF.head())
```

```python
#joining dataframes into single dateframe
PCE_PCEDG = pd.merge(PCE, PCEDG, on='observation_date', how='inner')
PCES_PCEF = pd.merge(PCES, PCEF, on='observation_date', how='inner')
PCE_joined = pd.merge(PCE_PCEDG, PCES_PCEF, on='observation_date', how='inner')
```

```python
print(PCE_joined.head())
```

```python
#saving joined dataframe as a csv in project folder
filepath = Path(folderpath + '\\PCE_Joined.csv')
filepath.parent.mkdir(parents=True, exist_ok=True)
PCE_joined.to_csv(filepath, index='false')
```
