This short scipt conducts some Exploratory Data Analysis view and export summary statistics, use Variance Inflation Factor to check for multicolinearity, and map relationships by determining the correlation between variables.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
folderpath = 'C:\\Users\\micha\\Desktop\\File References\\CIS480\\Project\\Data\\'
data = pd.read_csv(folderpath + 'Data.csv')

data.head()
```

```python
#Printing summary statistics and exporting it as a csv file
data.describe().to_csv(folderpath + 'Summary_Statistics.csv')
print(data.describe())
```

```python
#Calculating Variance Inflation Factor (VIF) to test for multicolinearity
X = data[['GDP_Growth', 'CCI']]
X_constant_added = sm.add_constant(X)

vif = vif = pd.DataFrame()
vif["VIF"] = [variance_inflation_factor(X_constant_added.values, i) for i in range(X_constant_added.shape[1])]
vif["feature"] = X_constant_added.columns
print(vif)
```

```python
#Mapping relationships
corr = data[['PCE', 'PCEDG', 'PCES', 'PCEF', 'GDP_Growth', 'CCI']].corr()
corr.to_csv(folderpath + 'Correlations.csv')
```
